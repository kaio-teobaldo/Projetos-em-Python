```python
def media(dados):
  dados = sorted(dados)
  n = len(dados)
  meio = n //2
  if n % 2 == 0:
    return (dados[meio - 1] + dados[meio])/2
  else:
    return dados[meio]
  print("Mediana:", mediana([1, 3, 2, 5, 4]))
```

```python
def moda(dados):
  cotagem = {}
  for valor in dados:
    contagem[valor] = contagem.get(valor, 0) + 1
    return max(contagem, key=contagem.get)
    print("Moda", moda([1, 2, 2, 3, 4]))
```

```python
def variancia(dados):
  m = sum(dados) / len(dados)
  return sum((x - m) ** 2 for x in dados) / len(dados)

print("Variancia:", variancia([2, 4, 6, 8]))
```

```python
import math
def desvio_padrao(dados):
  m = sum(dados)/ len(dados)
  var = sum((x - m)**2 for x in dados)/len(dados)
  return  var *0.5      #math.sqrt(var)
  print("Desvio padrão:", desvio_padrao([2, 4, 6, 8]))
```

```python
def amplitude(dados):
  return max(dados) - min(dados)
  print("Amplitude:", amplitude([10, 20, 30, 40]))
```

```python
def frequencia(dados):
    contagem = {}
    for valor in dados:
        contagem[valor] = contagem.get(valor, 0) + 1
    print("Frequencia", contagem)
frequencia([1, 2, 3, 3, 3])
```

```python
def percentual_acima_media(dados):
    m = sum(dados) / len(dados)
    acima = [x for x in dados if x > m]
    return (len(acima)/ len(dados)) * 100
print("Percentual acima da media:", percentual_acima_media([10, 20, 30, 40]))
```

```python
def procurar(frase, letra):
  indice = 0
  achou = False
  while indice < len(frase) and not achou:
    if frase[indice] == letra:
      achou = True
    else:
      indice = indice + 1 # Incrementa o índice se a letra não foi encontrada na posição atual
  if achou:
    return indice
  else :
    return -1

def main():
  frase = input("Digite uma frase:")
  caracter = input("Digite a letra que deseja procurar:")
  print("Indice:", procurar(frase, caracter))

main()
```
